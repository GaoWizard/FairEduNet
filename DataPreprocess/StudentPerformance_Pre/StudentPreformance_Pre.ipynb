```python
df = data
# 查看数据集的列类型
print("数据集的每列类型：")
print(df.dtypes)

# 获取分类数据类型的列
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
print("\n分类数据类型的特征：", categorical_columns)

# 统计分类特征的数量
num_categorical_features = len(categorical_columns)
print("\n分类特征的数量：", num_categorical_features)

# 查看数据集的总特征数量
num_features = df.shape[1]
print("\n数据集总共有", num_features, "个特征")
```

```python

# 统计每个分类特征的数量和百分比
summary = {}

# 遍历每个分类特征
for column in df.select_dtypes(include=['object', 'int64']).columns:
    counts = df[column].value_counts()  # 计算每个标签的数量
    percentages = df[column].value_counts(normalize=True) * 100  # 计算每个标签的百分比
    
    # 格式化输出每个特征的统计数据
    summary[column] = pd.DataFrame({
        'n': counts,
        'n (%)': counts.astype(str) + ' (' + percentages.round(2).astype(str) + '%)'
    })

# 输出每个特征的统计结果
for column, stats in summary.items():
    print(f"\n统计 {column} 列:")
    print(stats)
```

```python
df = data

# 统计每个分类特征的数量和百分比
summary = {}

# 遍历每个分类特征
for column in df.select_dtypes(include=['object', 'int64']).columns:
    counts = df[column].value_counts()  # 计算每个标签的数量
    percentages = df[column].value_counts(normalize=True) * 100  # 计算每个标签的百分比
    
    # 格式化输出每个特征的统计数据
    summary[column] = pd.DataFrame({
        'n': counts,
        'n (%)': counts.astype(str) + ' (' + percentages.round(2).astype(str) + '%)'
    })

# 将统计结果保存到 Excel 文件
with pd.ExcelWriter('stu_sta_result.xlsx') as writer:
    for column, stats in summary.items():
        # 将每个特征的统计数据存储到不同的工作表中
        stats.to_excel(writer, sheet_name=column, index=True)

print("Excel 文件已保存。")
```

```python
# 数据预处理
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
# 初始化标签编码器和独热编码器

# 二元属性使用标签编码，多个类别的属性使用独热编码

label_encoder = LabelEncoder()

onehot_encoder = OneHotEncoder(sparse=False, drop='first')  # drop='first' 以避免多重共线性
```

```python
# 选择需要标签编码的列 (二元属性)
binary_columns = ['schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic']

# 性别属性是敏感属性，不做处理

# 对二元属性列应用标签编码    
for column in binary_columns:
    data[column + '_encoded'] = label_encoder.fit_transform(data[column]).astype('int64')
```

```python
# 选择需要独热编码的其他object类型的列
other_object_columns = ['school', 'address', 'famsize', 'Pstatus', 'reason', 'guardian','Mjob', 'Fjob']

# 对这些列应用独热编码并合并到原始DataFrame中
for column in other_object_columns:
    dummies = pd.get_dummies(data[column], prefix=column, drop_first=True).astype(int)
    data = pd.concat([data, dummies.astype('int64')], axis=1)
```

```python
# 检查一下新增加的列
new_columns = [col for col in data.columns if '_encoded' in col or any(col.startswith(prefix) for prefix in other_object_columns)]
```
