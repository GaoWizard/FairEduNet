```python
import pandas as pd  # 用于数据处理和分析
import matplotlib.pyplot as plt  # 用于绘图
import seaborn as sns  # 用于绘制统计图形
```

```python
# 读取数据
file_path = 'C:/Code/StudentPerformance_Pre/stuper_pre.csv'
data = pd.read_csv(file_path)
```

```python
# 计算整个数据集的相关性矩阵
correlation_matrix = data.corr()

# 提取与G3_prediction相关的相关系数
g3_prediction_correlations = correlation_matrix["G3_prediction"].sort_values(ascending=False)

# 选择相关性最高的前10个特征
top_features = g3_prediction_correlations.index[1:11]

# 如果需要绘制热图，计算这些特征之间的相关性
top_correlations = data[top_features].corr()
```

```python
# 绘制热图
plt.figure(figsize=(10, 8))
sns.heatmap(top_correlations, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Top 10 Features Related to G3_prediction')
plt.show()
```

```python
# Plot the distribution of the top 10 features related to G3_prediction
top_features = g3_prediction_correlations.index[1:11]

data['sex'] = data['sex'].replace({0: 'female', 1: 'male'})
data['higher_encoded'] = data['higher_encoded'].replace({1: 'wants to take higher education', 0: "doesn't want to take higher education"})
data['Medu'] = data['Medu'].replace({0: 'none', 1: 'primary (4th)', 2: '5th-9th', 3: 'secondary', 4: 'higher'})
data['studytime'] = data['studytime'].replace({1: '<2 hours', 2: '2 to 5 hours', 3: '5 to 10 hours', 4: '>10 hours'})
data['Fedu'] = data['Fedu'].replace({0: 'none', 1: 'primary (4th)', 2: '5th-9th', 3: 'secondary', 4: 'higher'})
data['address_U'] = data['address_U'].replace({1: 'urban', 0: 'rural'})
data['internet_encoded'] = data['internet_encoded'].replace({1: 'yes', 0: 'no'})
data['reason_reputation'] = data['reason_reputation'].replace({1: 'yes', 0: 'no'})
data['Mjob_teacher'] = data['Mjob_teacher'].replace({1: 'yes', 0: 'no'})

# Set up the matplotlib figure
plt.figure(figsize=(15, 20))

# Create a subplot for each feature
for i, feature in enumerate(top_features, 1):
    plt.subplot(5, 2, i)
    sns.histplot(data, x=feature, hue="sex", multiple="stack", kde=True)
    plt.title(f'Distribution of {feature} by Sex')

# Adjust the layout
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))

# 使用两个相关性较高的特征绘制散点图
sns.scatterplot(data=data, x='G1', y='G2', hue='sex')
plt.title('Scatter Plot of G1 vs G2 by Sex')
plt.show()
```
