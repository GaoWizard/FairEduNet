```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 读取CSV文件
data = pd.read_csv('oulad.csv')
```

```python
# 按gender分组
male_data = data[data['gender'] == 0]['final_result']
female_data = data[data['gender'] == 1]['final_result']
```

```python
# 创建直方图
plt.figure(figsize=(10, 6))

# 绘制男性的直方图
plt.hist(male_data, bins=10, alpha=0.5, label='Male', color='blue', edgecolor='black')

# 绘制女性的直方图
plt.hist(female_data, bins=10, alpha=0.5, label='Female', color='pink', edgecolor='black')

# 添加标题和标签
plt.title('Distribution of Final Results by Gender')
plt.xlabel('Final Result')
plt.ylabel('Frequency')

# 添加图例
plt.legend()

# 显示图形
plt.show()
```
